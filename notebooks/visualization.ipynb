```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

We make some changes to the dataframe to make it easier to visualise.

```python
df = pd.read_pickle('../data/cardio_risk.pkl')
df.head()
```

```python
def change_education(education):

    if education == 1:
        return 'Basic Education'
    elif education == 2:
        return 'Intermediate Education'
    elif education == 3:
        return 'Advanced Education'
    else:
        return 'Higher Education'
```

```python
df['education'] = df['education'].apply(change_education)
```

```python
df['sex'] = df['sex'].apply(lambda sex: 'Male' if sex == 'M' else 'Female')
```

```python
df['is_smoking'] = df['is_smoking'].apply(lambda low: low.lower().capitalize())
```

```python
binary_list = ['BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes', 'TenYearCHD']
```

```python
for col in binary_list:
    df[col] = df[col].apply(lambda num: 'Yes' if num == 1 else 'No')
```
